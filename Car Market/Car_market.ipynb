
# k-NN Algorithm Predict Car Prices


```python
import pandas as pd
import numpy as np

columns = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 
        'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 
        'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']

cars = pd.read_csv('dataset/imports-85.data', names=columns)

cars.head()
```

```python
print(cars.shape)
cars.info()
```

```python
cars = cars.replace("?", np.nan)

cars.head()
```

```python
cars.info()
```

```python
cars[['normalized-losses', 'horsepower', 'peak-rpm', 'price']] = cars[['normalized-losses', 'horsepower', 'peak-rpm', 'price']].astype(float) 

cars.info()
```

```python
cars['num-of-cylinders'].value_counts()
```

```python
number_rename = {
    'four': 4,
    'six': 6,
    'five': 5,
    'eight': 8,
    'two': 2,
    'three': 3,
    'twelve': 12,
    4:4,
    6:6,
    5:5,
    8:8,
    2:2,
    3:3,
    12:12
}

cars['num-of-cylinders'] = (cars['num-of-cylinders'].map(number_rename)).astype(float)

cars.info()
```

```python
cars['num-of-doors'].value_counts()
```

```python
cars['num-of-doors'] = (cars['num-of-doors'].map(number_rename)).astype(float)
cars.info()
```

```python
# Missing_data
cars.isnull().sum()
```

```python
cars = cars.dropna(subset=['price'])

cars.isnull().sum()
```

```python
cars = cars.fillna(cars.mean())

cars.isnull().sum()
```

```python
num_features = ['symboling', 'normalized-losses', 'num-of-doors',
               'wheel-base', 'length', 'width', 'height', 'curb-weight',
               'num-of-cylinders', 'engine-size', 'compression-rate',
               'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg',
               'price']

cars2 = cars[num_features]

cars2.head()

cars2.isnull().sum()
```

```python
price_column = cars2['price']

#X_norm = (X-Xmin)/(Xmax - Xmin)
normalized = (cars2 - cars2.min())/(cars2.max() - cars2.min())

normalized['price'] = price_column

normalized.head()
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.neighbors import KNeighborsRegressor

def knn_train_test(train_col, target_col, df):
    
    np.random.seed(3)
    
    shuffle = np.random.permutation(df.index)
    df = df.reindex(shuffle)
    
    train_df = df.iloc[:101]
    test_df = df.iloc[101:]
    
    knn = KNeighborsRegressor()
    knn.fit(train_df[[train_col]], train_df[target_col])
    
    predictions = knn.predict(test_df[[train_col]])
    rmse = (mean_squared_error(test_df[target_col], predictions))**0.5
    
    return rmse  
```

```python
list_of_cols = normalized.columns.tolist()

list_of_cols.remove('price')

list_of_cols
```

```python
rmse_dict = {}

for each in list_of_cols:
    rmse = knn_train_test(each, 'price', normalized)
    rmse_dict[each] = rmse 
    
rmse_dict
rmse_series = pd.Series(rmse_dict)

rmse_series.sort_values(ascending = True)
```
